```python
import numpy as np              # Sayısal işlemler ve rastgele sayı üretimi için NumPy kütüphanesi
import pandas as pd             # Tablo (DataFrame) oluşturmak ve düzenlemek için Pandas kütüphanesi

OGRENCI_NO = 250401018          # Öğrenci numarası tanımlanır

random_seed = OGRENCI_NO % 97   # Öğrenci numarasının 97’ye göre modu alınarak random seed belirlenir
rng = np.random.default_rng(random_seed)  # Belirlenen seed ile rastgele sayı üreteci oluşturulur

n = int(rng.integers(8, 13))    # 8 ile 12 arasında (dahil) rastgele senaryo sayısı üretilir

engel_sayisi = rng.integers(5, 51, size=n)   # Her senaryo için 5–50 arasında rastgele engel sayısı üretilir
epsilon = rng.uniform(0.2, 2.0, size=n)      # Her senaryo için 0.2–2.0 arasında epsilon değerleri üretilir
kappa = rng.uniform(0.0, 10.0, size=n)       # Her senaryo için 0–10 arasında kappa değerleri üretilir

sapma_a = 0.7 * kappa + 0.3 * (1.0 / epsilon)  # Verilen formüle göre Sapma A hesaplanır
sapma_b = 0.4 * kappa + 0.2 * (1.0 / epsilon)  # Verilen formüle göre Sapma B hesaplanır

df = pd.DataFrame({            # Tüm veriler bir tablo (DataFrame) haline getirilir
    "Senaryo": np.arange(1, n+1),   # Senaryolar 1’den n’e kadar numaralandırılır
    "Engel_Sayisi": engel_sayisi,   # Engel sayıları tabloya eklenir
    "epsilon": epsilon,             # Epsilon değerleri tabloya eklenir
    "kappa": kappa,                 # Kappa değerleri tabloya eklenir
    "Sapma_A": sapma_a,             # Hesaplanan Sapma A değerleri tabloya eklenir
    "Sapma_B": sapma_b              # Hesaplanan Sapma B değerleri tabloya eklenir
})

print("random_seed =", random_seed, "| senaryo sayısı =", n)  # Kullanılan seed ve senaryo sayısı ekrana yazdırılır
print(df.round(4))               # Tablo ekrana yazdırılır ve ondalık sayılar 4 basamağa yuvarlanır
```
